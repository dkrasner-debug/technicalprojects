```python
#Importing Python packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
NoDupes = pd.read_csv('PyUploadPractice.csv')
NoDupes
```

```python
NoDupes.info()
```

```python
NoDupes.describe()
```

```python
pivot_table = NoDupes.pivot_table(index='attrition_reason_bucket',
                                        values= ('START_END_duration_days',
                                                 'START_END_duration_weeks',
                                                 'START_END_duration_months',
                                                 'START_END_duration_years'),
                                        aggfunc= 'median')
pivot_table
```

```python
pivot_table.to_excel("START_END_median_duration.xlsx",index = True)
```
